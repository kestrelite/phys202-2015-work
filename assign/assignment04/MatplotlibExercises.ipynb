```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter plots

Learn how to use Matplotlib's `plt.scatter` function to make a 2d scatter plot.

* Generate random data using `np.random.randn`.
* Style the markers (color, size, shape, alpha) appropriately.
* Include an x and y label and title.

```python
plt.plot(range(1, 11), np.random.randn(10), '.:')
plt.xlabel("Abitrary Data Point")
plt.ylabel("Arbitary Value")
plt.title("Arbitrary Data Points at Arbitrary Values")
```

## Histogram

Learn how to use Matplotlib's `plt.hist` function to make a 1d histogram.

* Generate randpom data using `np.random.randn`.
* Figure out how to set the number of histogram bins and other style options.
* Include an x and y label and title.

```python
plt.hist(np.random.randn(6000), bins=20, histtype='barstacked', cumulative=True)
plt.xlabel("Arbitrary Histogram")
plt.ylabel("Arbitrary Frequency")
plt.title("Arbitrary Cumulative Histogram of Arbitrary Data")
```
