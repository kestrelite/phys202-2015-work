This was a diagram used to indicate how gas flow near the base of the reactor changes the heat distribution in what's known as the divertor region. This diagram clearly demonstrates how the heat distribution changes over gas flow, by specifying exactly what the gas flow is. 

There are a few conventions in here that ther scales are based on. First, the scales are written in meters. This is always the case, and consequently this actually _does_ exhibit excellence in this regard. Having identified that the scale of meters was unnecessary, it was purposefully omitted. The target audience already knows the scale and its metrics. 

Second, for the most part, the actual data are given a meaningful scale. The scale allows the viewer to clearly see the heat discrepancy. While it does obscure what the hottest point actually is, this is again fine, as it occurs at a known point in the plasma along the separatrix.
